```python
import numpy as np
from sklearn import decomposition,datasets
```

```python
iris = datasets.load_iris()

X=iris['data']
model = decomposition.PCA(2)
model.fit(X)
Maxcomponent = model.components_  #最大特征向量，
ratio=model.explained_variance_ratio_
score=model.score(X)

print('返回具有最大方差的成分:',Maxcomponent)
print('保留主成分的方差贡献率:',ratio)
print('所有样本的log似然平均值:',score)
```

```python
import matplotlib.pyplot as plt

X_new = model.fit_transform(X)
g1=plt.figure(1,figsize=(8,6))
plt.scatter(X_new[:,0],X_new[:,1],c='r',cmap=plt.cm.Set1, edgecolor='k', s=40)#因为pca降维后是2维度，所以使用二维可视化现实
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('After the dimension reduction')
plt.show()
```
